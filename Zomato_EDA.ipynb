### Replace the Country Code with Country Name
- In the dataset, the country code column contains country codes corresponding to their respective countries. Replace the country codes with country names for better understanding.

```python
df['Country Code'].values  #values in the country code feature
```

```python
# import the excel which having the country name and correspoding country code
df_country=pd.read_excel("Country-Code.xlsx") 
```

```python
df_country
```

```python
df_country.shape
```

```python
#merge zomato dataset and country dataset based on countrycode
final_data=pd.merge(df,df_country,on="Country Code",how="left")
```

```python
final_data.shape
```

```python
final_data.columns
```

```python
final_data.head()
```

```python
final_data.dtypes
```

```python
#How many values are there for each country
final_data.Country.value_counts()
```

- Observation: India has the highest number of Zomato orders in the given dataset, followed by the United States

```python
country_names=final_data.Country.value_counts().index  #store the country names in the variable
```

```python
country_names
```

```python
country_values=final_data.Country.value_counts().values #store the zomato orders of the country.
```

```python
country_values
```

```python
#plot the country names and the corresponding the zomato orders
plt.pie(country_values,labels=country_names)
plt.show()
```

- Observation: The pie chart is not effective in displaying all the countries with their respective Zomato orders.


```python
#plot top 3 countries which have highest number of zomato order
plt.pie(country_values[:3],labels=country_names[:3],autopct="%1.2f%%")
plt.show()
```

### From the above pie chart, we can conclude that Zomato has more scope in India. The majority of Zomato orders come from India, followed by the US and the UK

```python
#combine some feature into single dataset
final_data.groupby(["Aggregate rating","Rating color","Rating text"]).size()
```

```python
#combine some feature into single dataset
final_data.groupby(["Aggregate rating","Rating color","Rating text"]).size().reset_index()
```

```python
#combine some feature into single dataset
ratings=final_data.groupby(["Aggregate rating","Rating color","Rating text"]).size().reset_index().rename(columns={0:"Rating Count"})
```

```python
ratings
```

### Observations:
- A rating between 4.5 and 4.9 is considered excellent.
- A rating between 4.0 and 4.4 is considered very good.
- A rating between 3.5 and 3.9 is considered good.
- A rating between 2.5 and 3.4 is considered average.
- A rating below 2.5 is considered poor.
- Most of the restaurants have a good rating.

### Plot the bar graph for the Rating

```python
import matplotlib
matplotlib.rcParams["figure.figsize"]=(12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```
