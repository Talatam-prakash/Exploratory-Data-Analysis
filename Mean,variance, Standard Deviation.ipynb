## Statistical Analysis

- import Libraries

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
iris = sns.load_dataset("iris")
```

```python
setosa = iris.loc[iris['species']=='setosa']
virginica = iris.loc[iris['species']=='virginica']
versicolor = iris.loc[iris['species']=='versicolor']
```

```python
print("Mean of Petal Length for each specices :")
print("Setosa      : ",np.mean(setosa['petal_length']))
print("Virginica   : ",np.mean(virginica['petal_length']))
print("Versicolor  : ",np.mean(versicolor['petal_length']))
```

- for the setosa the mean or average value for the distribution is the 1.46 and other so on 

```python
print("mean is change when the outlier is added")
print(np.mean(np.append(setosa['petal_length'],50)))
```

- clearly see that mean value is shifted 


## Standard Deviation is how wider the values are spreading

```python
print("Standard deviation of Petal Length for each specices :")
print("Setosa      : ",np.std(setosa['petal_length']))
print("Virginica   : ",np.std(virginica['petal_length']))
print("Versicolor  : ",np.std(versicolor['petal_length']))
```

# Median ---- middle value of the values in sorted order

```python
print("Median of Petal length for each species :")
print("Setosa      : ",np.median(setosa['petal_length']))
print("Virginica   : ",np.median(virginica['petal_length']))
print("Versicolor  : ",np.median(versicolor['petal_length']))
print("Median is not change when the outliers is added")
print("Setosa      : ",np.median(np.append(setosa['petal_length'],150)))
print("Virginica   : ",np.median(np.append(virginica['petal_length'],150)))
print("Versicolor  : ",np.median(np.append(versicolor['petal_length'],150)))
```

# Percentaile
- Percentaile means when the values in sorted order the 10th percentaile value indicates that there are 10 % values less than that value and 90% values greater than that values.
- 50th percentaile means median and 100th percentaile value means maximumum

```python
print(np.percentile(iris["petal_length"],50))
print(np.percentile(iris["petal_length"],np.arange(0,101,25)))
print(np.percentile(setosa["petal_length"],np.arange(0,101,25)))
```

- 50% percentage of the datapoints in the petal length are less than 4.35

# Box Plot

```python
sns.boxplot(x='species',y='petal_length',data=iris)
```

- for the versicolor the box start with nearly 4.1 means 25th percentile and ends with 4.7 means 75th percentile

- Setosa and versicolor having outlier

# Violin Plot
- Violin plot same as the box plot but have spreading of datapoints shows in leaf width

```python
sns.violinplot(x='species',y='petal_length',data=iris)
```

Violin plot have box plot inside the leaf and pdf of the wide spread of the leaf.
- black box inside the violin plot incidates the box plot with white spot is 50th percentile and edge of the box is 25th percentile and 75th percentile
- wide spread of the leafs is pdf od petal_length of the species

# Joint plot
- Combination of scatter plot and bar plot

```python
sns.jointplot(x='petal_length',y="petal_width",data=iris)
```

```python
sns.jointplot(x='petal_length',y="petal_width",data=setosa,kind='kde')
```

Joint plot represents the bivariant analysis of the attribute.
- at the value of petal_length 1.4 and petal_width 0.2 more circle means in that area the more values are plotted
- 1D density is PDF
- 2D density plot is contour plot like joint plot
