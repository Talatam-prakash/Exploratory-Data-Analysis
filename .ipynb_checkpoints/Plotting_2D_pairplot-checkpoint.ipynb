## Explore the Iris Dataset 


- Import the libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris=sns.load_dataset("iris")
```

```python
iris.head()
```

```python
iris.shape
#150 rows(datapoints) and 5 columns(features) 
```

```python
#what are the features in the dataset 
iris.columns
```

```python
#want specific count of each class in target class
iris['species'].value_counts()
```

```python
#2D plot for sepal length and sepal width
iris.plot(kind="scatter",x="sepal_length",y="sepal_width")
plt.show()
```

```python
sns.FacetGrid(iris,hue="species")\
.map(plt.scatter,"sepal_length","sepal_width")\
.add_legend();
plt.show()
```

```python
sns.scatterplot(x="sepal_length",y="sepal_width",data=iris,hue="species")
```

```python
sns.scatterplot(x="petal_length",y="petal_width",data=iris,hue="species")
```

 we have a clear difference in species when we plot petal length and petal width
species is easily separable

```python
#Draw Pairplot to show the different plots between features which have numerical values
sns.pairplot(iris,hue="species")
```

- x = petal width and y = petal length gives the clear seperation of species 
- petal_length < 2 and petal_width < 1   => then it is setosa
- petal widh >1 and petal length < 5 => versicolor
- otherwise virginca

#### Pair is working for the small dataset only

# Histogram

```python
sns.histplot(iris.petal_length)
```

```python
fig=plt.figure(figsize=(20,20))
sns.FacetGrid(iris,hue="species")\
.map(sns.histplot,"petal_length")\
.add_legend();
plt.show()
```
