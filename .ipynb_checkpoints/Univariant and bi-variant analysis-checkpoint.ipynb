# Cummalative Distribution Function and Probability Distribution Function

Pdf-- Probability Distribution Function is how many values for the certain feature at the certain values 
- like at petal length 4 there 10 values for setosa

CDF - Cummalative Distribution Function is how many values less than certain values for the specific point of values
- Eg : at petal length 4 the cdf is .84 for the setosa means the 84 % of the setosa values are less than 4 petal length

```python
plt.hist(setosa["petal_length"])
```

```python
counts,bin_edges=np.histogram(setosa["petal_length"],bins=10,density=True)
pdf=counts/(sum(counts))
print(counts)
print(sum(counts))
print(pdf)
print(bin_edges)
```

```python
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,label="pdf")
plt.plot(bin_edges[1:],cdf,label="cdf")
plt.legend()
plt.show()
```

```python
counts,bin_edges=np.histogram(virginica["petal_length"],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)

cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.show()
```

```python
counts,bin_edges=np.histogram(setosa["petal_length"],bins=10,density=True)
pdf=counts/(sum(counts))
print("setosa")
print(pdf)
print(bin_edges)

cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,label="setosa-pdf")
plt.plot(bin_edges[1:],cdf,label="setosa-cdf")
# plt.show()

#for iris_virginica
counts,bin_edges=np.histogram(virginica["petal_length"],bins=10,density=True)
pdf=counts/(sum(counts))
print("virginica")
print(pdf)
print(bin_edges)

cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,label="virginca-pdf")
plt.plot(bin_edges[1:],cdf,label="virginca-cdf")
# plt.show()

#for iris_versicolor
counts,bin_edges=np.histogram(versicolor["petal_length"],bins=10,density=True)
pdf=counts/(sum(counts))
print("versicolor")
print(pdf)
print(bin_edges)

cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,label="versicolor-pdf")
plt.plot(bin_edges[1:],cdf,label="versicolor-cdf")
plt.grid()
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
plt.show()
```

By the above plots the setosa is petal length less than 2, versicolor is petal length greater than 3 and less than 5 and virginica is petal length greater than 5.
